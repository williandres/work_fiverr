```
import openai
import csv


openai.api_key = "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"

riddles = []
for _ in range(50):
    prompt = "Create a unique riddle, separating the riddle and its answer with '|'. Remember to be creative and distinct from previous prompts:"
    respuesta = openai.Completion.create(
        prompt=prompt,
        max_tokens=100,
        model="text-davinci-003",
        temperature=0.8,
        top_p=1,
        frequency_penalty=0.8,
        presence_penalty=0.8
    )
    print(respuesta.choices[0].text.strip())
    row = respuesta.choices[0].text.strip().split("|")
    row = [elemento.strip() for elemento in row]

    riddles.append(row)

with open("riddles.csv", mode="w", newline="", encoding="utf-8") as archivo_csv:
    writer = csv.writer(archivo_csv)
    writer.writerow(["Riddles", "Answers"])
    for row in riddles:
        writer.writerow(row)
```

```
import openpyxl
import csv

matrix = []
with open("riddles_base.csv", mode="r", newline="", encoding="utf-8") as archivo:
    lector_csv = csv.reader(archivo)
    next(lector_csv)

    for fila in lector_csv:
        matrix.append(["{","","",""])
        matrix.append(["","'question':'",fila[0],"',"])
        matrix.append(["","'answer':'",fila[1],"',"])
        matrix.append(["},","","",""])

with open("riddles.csv", mode="r", newline="", encoding="ISO-8859-1") as archivo:
    lector_sv = csv.reader(archivo)
    next(lector_sv)

    for fila in lector_sv:
        matrix.append(["{","","",""])
        matrix.append(["","'question':'",fila[0],"',"])
        matrix.append(["","'answer':'",fila[1],"',"])
        matrix.append(["},","","",""])

# Crear un nuevo libro de Excel
libro = openpyxl.Workbook()
hoja = libro.active

# Escribir la matriz en la hoja
for fila in matrix:
    hoja.append(fila)

# Guardar el archivo de Excel
nombre_archivo = "riddles.xlsx"
libro.save(nombre_archivo)
```
